```python
import pandas as pd
import numpy as np
```

```python
# Load data

df = pd.read_csv('AnnoMI-full-cleaned-vggish-embedded.csv')
```

```python
df.head()
```

```python
# Make a dataframe where the interlocutor is therapist
df_therapist = df[df['interlocutor'] == 'therapist']
df_therapist.head()
```

```python
# Split the data uniquely based on vidoe_title

video_urls = df_therapist['video_url'].unique()

# Randomly shuffle the video URLs
np.random.shuffle(video_urls)

# Split the video URLs into training and test sets
# 80% of the data goes to the training set, and 20% goes to the test set
split_index = int(len(video_urls) * 0.7)  
train_video_urls = video_urls[:split_index]  # First 80% of video URLs
test_video_urls = video_urls[split_index:]  # Remaining 20% of video URLs

# Split the data into training and test sets
df_train_therapist = df_therapist[df_therapist['video_url'].isin(train_video_urls)]
df_test_therapist = df_therapist[df_therapist['video_url'].isin(test_video_urls)]
```

```python
# make the x and y data

X_train = df_train_therapist['therapist_vggish_emb']
y_train = df_train_therapist['main_therapist_behaviour']

X_test = df_test_therapist['therapist_vggish_emb']
y_test = df_test_therapist['main_therapist_behaviour']
```
