```python
# Count the number of time client_talk_type is 'neutral'
neutral = df['client_talk_type'].value_counts()['neutral']

# Count the number of time client_talk_type is 'sustain'
sustain = df['client_talk_type'].value_counts()['sustain']

# Count the number of time client_talk_type is 'change'
change = df['client_talk_type'].value_counts()['change']

# Now make ratios of the counts for neutral 

result_n = neutral / (neutral + sustain + change)
result_s = sustain / (neutral + sustain + change)
result_c = change / (neutral + sustain + change)

# Print the results
print("Neutral: ", result_n)
print("Sustain: ", result_s)
print("Change: ", result_c)
```

```python
print(neutral + sustain + change)
print(neutral)
print(sustain)
print(change)
```

```python
# Make two data frames, one where mi_quality is High and one where mi_quality is Low
df_mi_high = df[df['mi_quality'] == 'high']
df_mi_low = df[df['mi_quality'] == 'low']
```

```python
# Plot client_talk_type in histogram for both data frames in one plot with legend
# high as lightgreen and low as purple
import matplotlib.pyplot as plt

df_mi_high['client_talk_type'].value_counts().plot(kind='bar', color='lightgreen')
df_mi_low['client_talk_type'].value_counts().plot(kind='bar', color='purple')
plt.legend(['high', 'low'])
```

```python
# Same for main_therapist_behaviour
df_mi_high['main_therapist_behaviour'].value_counts().plot(kind='bar', color='lightgreen')
df_mi_low['main_therapist_behaviour'].value_counts().plot(kind='bar', color='purple')
plt.legend(['high', 'low'])
```
